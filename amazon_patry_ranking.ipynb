## Data Cleaning before modeling

```python
#we have removed empty rank products
import numpy as np
pantry_meta["length"]=pantry_meta["rank"].str.len()
pantry_meta
pantry_meta_clean=pantry_meta[pantry_meta["length"]!=0]
pantry_meta_clean.shape
pantry_meta_clean.head(2)
```

```python
#Cleaning rank column
import re
r="([0-9]*,*[0-9]*)\s"
pantry_meta_clean["rank_clean"]=pantry_meta_clean["rank"].str.extract(r)
pantry_meta_clean["rank_clean"]=pantry_meta_clean["rank_clean"].str.replace(',','')
pantry_meta_clean["rank_clean"]=pantry_meta_clean["rank_clean"].astype(int)
pantry_meta_clean.head(2)
```

```python
main_data=pd.merge(left=pantry_meta_clean,right=pantry,left_on="asin",right_on="asin")
```

```python
#Preparing a final merged dataframe that has all features of our interest- average rating, total reviews, crowd puller score
product_wise_rating_data=pantry.groupby("asin")[["overall"]].mean()
product_wise_reviews_data=pantry.groupby("asin")[["reviewerID"]].count()
X_pre_train=pd.merge(left=product_wise_reviews_data,right=product_wise_rating_data,left_on="asin",right_on="asin")
pre_train=pd.merge(left=X_pre_train,right=pantry_meta_clean,left_on="asin",right_on="asin")[["asin","overall","reviewerID","rank_clean","Crowd puller"]]
pre_train=pre_train.rename(columns={"reviewerID":"total_reviews","overall":"avg_rating"})
pre_train["ratingXreview"]=pre_train["avg_rating"]*pre_train["total_reviews"]
```

```python
#Price Cleaning to include price as another feature of interest
pantry_meta_clean["price"]=pantry_meta_clean["price"].str.strip()
pantry_meta_clean["price"]=pantry_meta_clean["price"].str.replace('$','')
pantry_meta_cleanprice=pantry_meta_clean[pantry_meta_clean["price"].str.len()>0]
pantry_meta_cleanprice=pantry_meta_cleanprice[~pantry_meta_cleanprice["price"].str.contains('amazon')]
pantry_meta_cleanprice["price"]=pantry_meta_cleanprice["price"].astype(float)
price_data=pantry_meta_cleanprice[["asin","price"]]
```

```python
pre_train=pd.merge(left=pre_train,right=price_data,left_on="asin",right_on="asin")
```

```python
#removing outliers
def remove_outliers(data, variable, lower=-np.inf, upper=np.inf):
    """
    Input:
      data (DataFrame): the table to be filtered
      variable (string): the column with numerical outliers
      lower (numeric): observations with values lower than or equal to this will be removed
      upper (numeric): observations with values higher than this will be removed
    Output:
      a DataFrame with outliers removed
    Note: This function should not change mutate the contents of data.
    """
    return data[(data[variable]>lower) & (data[variable]<=upper)]
```

```python
def rmse(predicted, actual):
    """
    Calculates RMSE from actual and predicted values
    Input:
      predicted (1D array): vector of predicted/fitted values
      actual (1D array): vector of actual values
    Output:
      a float, the root-mean square error
    """
    return np.sqrt(np.mean((actual - predicted)**2))
```
